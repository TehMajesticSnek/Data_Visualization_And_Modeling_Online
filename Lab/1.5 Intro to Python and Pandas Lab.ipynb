```python
# Import Pokemon Dataset
import pandas as pd
df = pd.read_csv('pokemon.csv')
df.head()
```

># What is the most common pokemon type?

```python
df[['Type I', 'Type II']].stack().value_counts().plot(kind='bar')
```

Water types are the most common, followed by normal, flying grass, and psychic.

># Which pokemon rank the highest in each base stat, and total?

```python
sort = df.sort_values('HP', ascending=False)
sort[['National\nDex', 'Pokemon\nName', 'HP']].head()
```

```python
sort = df.sort_values('Attack', ascending=False)
sort[['National\nDex', 'Pokemon\nName', 'Attack']].head()
```

```python
sort = df.sort_values('Special\nAttack', ascending=False)
sort[['National\nDex', 'Pokemon\nName', 'Special\nAttack']].head()
```

```python
sort = df.sort_values('Defense', ascending=False)
sort[['National\nDex', 'Pokemon\nName', 'Defense']].head()
```

```python
sort = df.sort_values('Special\nDefense', ascending=False)
sort[['National\nDex', 'Pokemon\nName', 'Special\nDefense']].head()
```

```python
sort = df.sort_values('Speed', ascending=False)
sort[['National\nDex', 'Pokemon\nName', 'Speed']].head()
```

```python
sort = df.sort_values('Total', ascending=False)
sort[['National\nDex', 'Pokemon\nName', 'Total']].head()
```

># Which abilities are most common and most rare?

```python
df[['Ability I', 'Ability II', 'Hidden Ability']].stack().value_counts().head(20).plot(kind='bar')
```

```python
df[['Ability I', 'Ability II', 'Hidden Ability']].stack().value_counts().tail(20).plot(kind='bar')
```

Swift Swim is most common, followed by Intimidate, Sturdy, levitate, and Chlorophyll.\
There are many abilities unique to specific pokemon, so no one most rare

># Which pokemon are hardest to catch?

```python
sort = df.sort_values('Catch Rate', ascending=False)
sort[['National\nDex', 'Pokemon\nName', 'Catch Rate']].tail(30)
```

A few pokemon have no listed rate as they aren't caught traditionally.  
Otherwise, most legendary and mythical pokemon share the lowest spot with a catch rate of 3.0 (out of 255).  
*Many not listed
