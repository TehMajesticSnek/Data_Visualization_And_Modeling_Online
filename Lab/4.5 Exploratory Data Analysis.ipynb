```python
import pandas as pd
import matplotlib.pyplot as mlib
import seaborn as sns

df = pd.read_csv('StudentsPerformance.csv')

df.head()
```

I want to compare parental education to school performance. We'll be looking at 6 columns. Parent level of education, our target, indicates what level of schooling parents completed. This includes some high school, high school, some college, associate's degree, backelor's degree, and master's degree. Lunch shows whether students recieve reduced cost lunch, or standard. Test preparation course just shows whether a student has taken a prep-course or not, and the scores are of course what they got on standard tests.

```python
parent_edu = df['parental level of education']
parent_edu.value_counts().plot(kind='bar')
```

```python
group_ct = pd.crosstab(df['parental level of education'], df['lunch'])
group_ct.plot(kind='bar', stacked=False, title='Student Lunch Price Reduction Compared to Parent Education', ylabel='Count')
mlib.legend(title='Lunch Reduction')
```

```python
group_ct = pd.crosstab(df['parental level of education'], df['test preparation course'])
group_ct.plot(kind='bar', stacked=False, title='Test Preparation Course Completion Compared to Parent Education', ylabel='Count')
mlib.legend(title='Test Prep Course')
```

```python
sns.histplot(x='math score', data=df, hue='parental level of education', element="step")

mlib.title('Math Scores Compared to Parent Education')
mlib.xlabel('Score')
```

```python
sns.histplot(x='reading score', data=df, hue='parental level of education', element="step")

mlib.title('Reading Scores Compared to Parent Education')
mlib.xlabel('Score')
```

```python
sns.histplot(x='writing score', data=df, hue='parental level of education', element="step")

mlib.title('Writing Scores Compared to Parent Education')
mlib.xlabel('Score')
```

I was surprised to see very little if any correlation in this dataset. There are the typical 'smaller groups have less,' but there wasn't much else in terms of major trends.  
I suppose big picture, you have an equal chance at a successful education regardless of your parents' educational status.
